```python
import pandas as pd
from pycaret.classification import *
from sklearn.ensemble import IsolationForest
```

```python
inputs = 'inputs'
outputs = 'outputs'

train = pd.read_csv(inputs + '/train.csv')
test = pd.read_csv(inputs + '/test.csv')
submission = pd.read_csv(outputs + '/sample_submission.csv')
```

```python
def normalize_out(dataframe, remove_col):
    dff = dataframe
    for k in remove_col:
        level_1q = dff[k].quantile(0.1)
        level_3q = dff[k].quantile(0.9)
        IQR = level_3q - level_1q
        rev_range = 3  # 제거 범위 조절 변수
        dff[k][(dff[k] > level_3q + (rev_range * IQR))] = level_3q
        dff[k][(dff[k] < level_1q - (rev_range * IQR))] = level_1q
    return dff

train = normalize_out(train, ['introelapse', 'testelapse', 'surveyelapse', 'age', 'familysize'])
test = normalize_out(test, ['introelapse', 'testelapse', 'surveyelapse', 'age', 'familysize'])
```

```python
clf = setup(data=train, target='nerdiness')
```

```python
best_2 = compare_models(sort='AUC', n_select=2)
```

```python
blended = blend_models(estimator_list=best_2, fold=5, method='soft')
```

```python
final_model = finalize_model(blended)
```

```python
predictions = predict_model(final_model, test, raw_score=True)
predictions
```

```python
submission['nerdiness'] = predictions['Label']
submission
```

```python
submission.to_csv(outputs + '/pycaret_submission.csv', index=False)
```
